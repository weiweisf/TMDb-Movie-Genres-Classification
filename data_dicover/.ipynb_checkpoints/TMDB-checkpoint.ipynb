```python
from glob import glob
import pandas as pd
import numpy as np
import os
import json
from collections import defaultdict
```

```python
glob("dataset/*")
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
display(df.sort_values(by='release_date', ascending=False).head())
df_credits = pd.read_csv('dataset/tmdb_5000_credits.csv')
df_credits.head()
```

```python
# most frequent countries
production_countries = defaultdict(int)
for row in df.production_countries:
    row = json.loads(row)
    for entry in row:
        production_countries[entry['name']] += 1

for tup in sorted(production_countries.items(), key=lambda x: -x[1]):
    print(tup)
```

```python
# most frequent companies
production_companies = defaultdict(int)
for row in df.production_companies:
    row = json.loads(row)
    for entry in row:
        production_companies[entry['name']] += 1

for tup in sorted(production_companies.items(), key=lambda x: -x[1]):
    print(tup)
```

```python
# most frequent categories
categories = defaultdict(int)
for row in df.genres:
    row = json.loads(row)
    for entry in row:
        categories[entry['name']] += 1

for tup in sorted(categories.items(), key=lambda x: -x[1]):
    print(tup)
```

```python
# most frequent actors
actors = defaultdict(int)

for row in df_credits.cast:
    row = json.loads(row)
    for cast_member in row:
        actors[cast_member['name']] += 1

for tup in sorted(actors.items(), key=lambda x: -x[1]):
    print(tup)
```

```python
# most frequent spoken-language
languages = defaultdict(int)

for row in df.spoken_languages:
    row = json.loads(row)
    for cast_member in row:
        languages[cast_member['name']] += 1

for tup in sorted(languages.items(), key=lambda x: -x[1]):
    print(tup)
```
